```python
import os
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt
```

```python
# Directory where your CSV files are stored
directory = 'data'

# Step 1: Load and transpose each CSV file
song_data = []
song_names = []

for filename in sorted(os.listdir(directory)):
    if filename.endswith(".csv"):
        # Read and transpose CSV
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath, header=None).transpose()
        
        # Store the transposed data and song name
        song_data.append(df.values)  # Convert to numpy array for DTW
        song_names.append(filename.split('.')[0])  # Store song name
```

```python
# Step 2: Calculate DTW distance matrix
n_songs = 4
dtw_matrix = np.zeros((n_songs, n_songs))

for i in range(n_songs):
    for j in range(i+1, n_songs):  # Calculate only upper triangle (matrix is symmetric)
        distance, _ = fastdtw(song_data[i], song_data[j], dist=euclidean)
        dtw_matrix[i, j] = distance
        dtw_matrix[j, i] = distance  # Symmetric
        print(f'{i} and {j}')
```

```python
import seaborn as sns

plt.figure(figsize=(30, 30), dpi=500)
sns.heatmap(dtw_matrix, annot=True, fmt=".2f", cmap='viridis', 
            xticklabels=range(1, n_songs), yticklabels=range(1, n_songs))
plt.title("DTW Distance Matrix")
plt.xlabel("Songs")
plt.ylabel("Songs")
plt.show()
```
