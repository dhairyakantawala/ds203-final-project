```python
import os
import numpy as np
import pandas as pd
import librosa

def create_MFCC_csv(file_name, output_csv):
    sr_value = 44100
    n_mfcc_count = 20
    
    try:
        y, sr = librosa.load(file_name, sr=sr_value)
        
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc_count)
        
        coeff_df = pd.DataFrame(mfccs)
        
        coeff_df.to_csv(output_csv, index=False)
        
        print(f"MFCC coefficients saved as {output_csv}")
        
    except Exception as e:
        print(f"Error creating MFCC coefficients: {file_name}: {str(e)}")

input_file = "training_data/JANA GANA MANA | National Anthem | 2017.mp4"
output_csv = "data/jana-gana-mana-MFCC.csv"
create_MFCC_csv(input_file, output_csv)
```
